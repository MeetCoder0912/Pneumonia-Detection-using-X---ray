```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define constants
IMAGE_WIDTH = 150
IMAGE_HEIGHT = 150
BATCH_SIZE = 32

# Create data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

# Create data iterators
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/Kaggle/pneumonia/chest_xray/train',
    target_size=(IMAGE_WIDTH, IMAGE_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    '/content/drive/MyDrive/Kaggle/pneumonia/chest_xray/val',
    target_size=(IMAGE_WIDTH, IMAGE_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='binary')
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
import matplotlib.pyplot as plt

# Get a batch of images and labels from the generator
images, labels = next(train_generator)

# Plot the images
plt.figure(figsize=(16, 8))
for i in range(BATCH_SIZE):
    plt.subplot(4, 8, i + 1)
    plt.imshow(images[i])
    plt.title('PNEUMONIA' if labels[i] else 'NORMAL')
    plt.axis('off')
plt.show()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(validation_generator)
)
```

```python
import matplotlib.pyplot as plt

# Plot the training and validation accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

# Plot the training and validation loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
test_generator = test_datagen.flow_from_directory(
    '/content/drive/MyDrive/Kaggle/pneumonia/chest_xray/test',
    target_size=(IMAGE_WIDTH, IMAGE_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='binary')

test_loss, test_acc = model.evaluate(test_generator, steps=len(test_generator))
print('Test accuracy:', test_acc)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# Get the path to a test image
test_image_path = '/content/drive/MyDrive/Kaggle/pneumonia/chest_xray/test/PNEUMONIA/person100_bacteria_475.jpeg'

# Load and preprocess the image
img = image.load_img(test_image_path, target_size=(IMAGE_WIDTH, IMAGE_HEIGHT))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.

# Make a prediction
prediction = model.predict(img_array)

# Print the prediction
if prediction[0] > 0.5:
    print("Prediction: PNEUMONIA")
else:
    print("Prediction: NORMAL")

# Display the image
plt.imshow(img)
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from PIL import Image
import io
import numpy as np
from tensorflow.keras.preprocessing import image

# Create UI elements
uploader = widgets.FileUpload(
    accept='image/*',
    multiple=False,
    description='Upload X-ray'
)

button = widgets.Button(description="Predict")
output_area = widgets.Output()

def on_button_clicked(b):
    if not uploader.value:
        with output_area:
            clear_output()
            print("Please upload an image first.")
        return

    # Get the uploaded image data
    # The uploader.value is a dictionary, get the first file's content
    file_info = list(uploader.value.values())[0]
    uploaded_image = Image.open(io.BytesIO(file_info['content']))

    # Preprocess the image
    img = uploaded_image.resize((IMAGE_WIDTH, IMAGE_HEIGHT))
    if img.mode != 'RGB':
        img = img.convert('RGB')
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.

    # Make a prediction
    prediction = model.predict(img_array)
    confidence = prediction[0][0]

    # Display the result
    with output_area:
        clear_output()
        display(uploaded_image.resize((300, 300))) # Display a resized version for consistency
        if confidence > 0.5:
            print(f"Prediction: PNEUMONIA (Confidence: {confidence*100:.2f}%)")
        else:
            print(f"Prediction: NORMAL (Confidence: {(1-confidence)*100:.2f}%)")


button.on_click(on_button_clicked)

# Display the UI
display(uploader, button, output_area)
```

```python
from tensorflow.keras.applications import VGG16

# Load the VGG16 model, pre-trained on ImageNet, without the top classification layers
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, 3))

# Print the model summary
conv_base.summary()
```

```python
conv_base.trainable = False
```

```python
model = tf.keras.models.Sequential([
    conv_base,
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(validation_generator)
)
```

```python
import matplotlib.pyplot as plt

# Plot the training and validation accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

# Plot the training and validation loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
test_loss, test_acc = model.evaluate(test_generator, steps=len(test_generator))
print('Test accuracy:', test_acc)
```

# Pneumonia Detection from Chest X-ray Images

This project uses deep learning to classify chest X-ray images as either "NORMAL" or "PNEUMONIA". It explores two different convolutional neural network (CNN) architectures: a custom CNN built from scratch and a more advanced model using transfer learning with the pre-trained VGG16 model.

The project is implemented in a Google Colab notebook and includes an interactive UI for uploading and classifying your own X-ray images.

## Dataset

The dataset used for this project is the "Chest X-Ray Images (Pneumonia)" dataset available on Kaggle.

- **Dataset Link:** [https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia](https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia)

The dataset is organized into three folders: `train`, `test`, and `val`, and contains a total of 5,863 chest X-ray images.

## Models

Two different models were built and evaluated in this project:

1.  **Custom CNN:** A simple CNN built from scratch with three convolutional layers. This model served as a baseline.
2.  **VGG16 Transfer Learning:** A more advanced model that uses the VGG16 architecture, pre-trained on the ImageNet dataset. This approach leverages the powerful features learned by VGG16 and fine-tunes them for the pneumonia detection task, resulting in a more stable and accurate model.
